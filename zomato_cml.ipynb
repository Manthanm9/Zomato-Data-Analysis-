# Creating a function that suggest list of restaurants based on givven parameters


This function takes 3 parameters: **cost for 2 people**,**location** and **Restaurant type** and return the name of resturants available at that location.

Let's find the names of restaurants that fall in Whitefield for casual dining and have cost below Rs.1000??



```python
def find_restaurant(location,restaurant,x):
    budget = data[(data['approx_cost(for two people)'] <= x) & (data['location'] == location) & (data['rest_type'] == restaurant) & (data['rate'] > 4)]
    return (budget['name'].unique())
```

```python
find_restaurant('Whitefield','Casual Dining',1000)
```
